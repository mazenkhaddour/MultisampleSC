```python
input_files = []
output_file = 'concatenated.h5ad'
batch_col = 'sample'
```

## Setting Envs 

```python
import scanpy as sc
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

# Read all files 

```python
adatas = [sc.read_h5ad(f) for f in input_files]
```

```python
print(f"Loaded {len(adatas)} samples")
for i, adata in enumerate(adatas):
    print(f"Sample {i+1}: {adata.n_obs} cells, {adata.n_vars} genes")
```

## Concatenate all objects

```python
adata = sc.concat(adatas, join='outer' )
```

```python
print(f"\nConcatenated: {adata.n_obs} cells, {adata.n_vars} genes")
```

# Filtering and assigning raw layer

```python
adata
```

```python
adata.layers["raw"] = adata.X.copy()
```

# save concatenated object

```python
adata.write_h5ad(output_file)
```
